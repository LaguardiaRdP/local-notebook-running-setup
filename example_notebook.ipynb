**Setup:** replace the `.env.localrunner` with your credentials, then run cells top to bottom.

**To run a different query:** add/swap the `.sql` file path in cell 2.

**To force a fresh pull from BigQuery:** set `full_refresh=True` in cell 2.

```python
# Cell 1 — Load credentials from .env.localrunner (run this first)
import subprocess
import os

script_path = os.path.join(os.getcwd(), '.env.localrunner')
if os.path.isfile(script_path):
    with open(script_path, 'r') as f:
        for line in f.readlines():
            line = line.strip()
            if line and not line.startswith('#') and '=' in line:
                key, _, value = line.partition('=')
                key = key.strip()
                value = value.strip()
                # Remove surrounding quotes (single or double) from shell-style values
                if (value.startswith('"') and value.endswith('"')) or (value.startswith("'") and value.endswith("'")):
                    value = value[1:-1]
                os.environ[key] = value
    print('Credentials loaded from .env.localrunner')
else:
    print(f'.env.localrunner not found at {script_path} — fill it in first')
```

```python
# Cell 2 — Run a query
# Point this at any .sql file inside the queries/ folder.
# Set full_refresh=True to re-run against BigQuery and update the cache.
from bigquery_adapter import run_bq_query

df = run_bq_query('queries/example_query.sql', full_refresh=False)
```
